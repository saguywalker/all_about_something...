```python
def read_dataset(num):
    path = "mi_bci/subjects/s"
    f = []
    for i in range(1,num+1):
        f.append(sio.loadmat(path+str(i).zfill(2)+".mat")['eeg'])
    return f
```

```python
x1, x2 = [], []
for i, ff in enumerate(f):
    x_1, x_2 = [], []
    trial = [j for j, a in enumerate(ff[0][0][11][0]) if a == 1]
    for j in range(len(trial)-1):
        x_1.append(ff[0][0][7][:64,trial[j]:trial[j+1]].transpose())
        x_2.append(ff[0][0][8][:64,trial[j]:trial[j+1]].transpose())
    x1.append(np.asarray(x_1))
    x2.append(np.asarray(x_2))
```

```python
x1 = [np.asarray(x) for x in x1]
x2 = [np.asarray(x) for x in x2]
```

```python
for i, x in enumerate(x1):
    print(i,x.shape)
print('*'*80)
for i, x in enumerate(x2):
    print(i,x.shape)
```

```python
max1, max2, max3 = 0, 0, 0
for i,x in enumerate(x1):
    if x.shape[0] > max1:
        print(i,0,x.shape[0])
        max1 = x.shape[0]
    if x.shape[1] > max2:
        print(i,1,x.shape[1])
        max2 = x.shape[1]
    if x.shape[2] > max3:
        print(i,2,x.shape[2])
        max3 = x.shape[2]
```

```python
y1, y2 = [], []
for i, x in enumerate(x1):
    if x.shape != (max1,max2,max3):
        tmp = np.zeros((max1,max2,max3))
        tmp[:x.shape[0],:x.shape[1],:x.shape[2]] = x
        x1[i] = tmp
        tmp2 = np.zeros((max1))
        tmp2[:x.shape[0]] = [1] * x.shape[0]
    else:
        tmp2 = [1] * max1
    y1.append(tmp2)
        
for i, x in enumerate(x2):
    if x.shape != (max1,max2,max3):
        tmp = np.zeros((max1,max2,max3))
        tmp[:x.shape[0],:x.shape[1],:x.shape[2]] = x
        x2[i] = tmp
        tmp2 = np.zeros((max1))
        tmp2[:x.shape[0]] = [2] * x.shape[0]
    else:
        tmp2 = [2] * max1
    y2.append(tmp2)
```

```python
#x1 = np.asarray(x1)
#x2 = np.asarray(x2)
y1 = [np.asarray(y) for y in y1]
y2 = [np.asarray(y) for y in y2]
```

```python
for i, x in enumerate(x1):
    print(i,x.shape)
print('*'*80)
for i, x in enumerate(x2):
    print(i,x.shape)
print('*'*80)
for i, x in enumerate(y1):
    print(i,x.shape)
print('*'*80)
for i, x in enumerate(y2):
    print(i,x.shape)
```

```python
def write_npz_file(x,y,folder,head,subject):
    if not os.path.exists(folder):
        os.makedirs(folder)
    filename = str(folder)+"/S"+str(subject).zfill(2)+"_"+head+".npz"
    np.savez(filename,x=x,y=y)
```

```python
for i, x in enumerate(x1):
    write_npz_file(x1[i],y1[i],"mi_bci","left",i+1)
```

```python
for i, x in enumerate(x2):
    write_npz_file(x2[i],y2[i],"mi_bci","right",i+1)
```
