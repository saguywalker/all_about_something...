```python
import tensorflow as tf
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import mne
from mne.io import concatenate_raws, read_raw_edf
from mne.datasets import eegbci
from mne import Epochs, pick_types, find_events
import os
```

```python
#read eeg file functions

#read from mne dataset
def read_from_mne(n):
    raw_files = []
    all_cases = [i for i in range(1,15)]
    for i in range(1,n+1):
        if i == 88:
            continue
        for j in range(1,15):
            if j == 1 or j == 2:
                label = 0
            else:
                label = int((j+1)%4+1)
            raw_files.append([eegbci.load_data(i,all_cases),label])
    return raw_files


#read from local file
def read_from_local(n):
    raw_files = []
    for i in range(1,n+1):
        if i == 88:
            continue
        sub_folder = path_root+"S"+str(i).zfill(3)+"/S"+str(i).zfill(3)+"R"
        for j in range(1,15):
            full_path = sub_folder+str(j).zfill(2)+".edf"
            if j == 1 or j == 2:
                label = 0
            else:
                label = int((j+1)%4+1)
            raw_files.append([read_raw_edf(full_path,preload=True,stim_channel="auto"),label])
    return raw_files

#read from local file with only imagine
def read_imagine_from_local(n):
    raw_files = []
    for i in range(1,n+1):
        sub_folder = path_root+"S"+str(i).zfill(3)+"/S"+str(i).zfill(3)+"R"
        for j in range(4,15,4):
            full_path = sub_folder+str(j).zfill(2)+".edf"
            raw_files.append(read_raw_edf(full_path,preload=True,stim_channel="auto"))
    return raw_files

#read imagine from mne dataset
def read_from_mne(n):
    raw_files = []
    all_cases = [i for i in range(1,15)]
    for i in range(1,n+1):
        for j in range(4,15,4):
            raw_files.append(eegbci.load_data(i,all_cases))
    return raw_files
```

```python
#define path root 
path_root = "subjects/"
event_id = dict(T1=2,T2=3)
tmin, tmax = 0.0, 3.0
```

```python
#read edf from local file
raw_files = read_imagine_from_local(109)
print("length of raw_files:",len(raw_files))
```

```python
#from this, S88, 92, 99 have different sample rate
for i in range(len(raw_files)-1,-1,-1):
    if raw_files[i].info['sfreq'] != 160:
        print(i,raw_files[i].info['sfreq'])
        del raw_files[i]

#print length after delete
print("length of raw_files:",len(raw_files))
```

```python
#fetch events from each raw_files
events = [ np.sort(find_events(r, shortest_event=0, stim_channel='STI 014'),axis=0) for r in raw_files]
```

```python
picks = pick_types(raw_files[0].info, meg=False, eeg=True, stim=False, eog=False,exclude='bads')
picks
```

```python
#pick only events T1(2) and T2(3)
events23 = []
events23_id = []
for e in events:
    e23 = []
    e23_id = []
    for i in range(len(e)):
        if e[i,-1] == 2 or e[i,-1] == 3:
            e23.append(e[i])
            e23_id.append(e[i,-1])
    events23.append(e23)
    events23_id.append(e23_id)
```

```python
data = []
labels = []
for i in range(len(raw_files)):
    epochs = Epochs(raw_files[i], events23[i],event_id,tmin,tmax, proj=True, picks=picks,baseline=None, preload=True)
    data.append(epochs.get_data())
    labels.append(events23_id[i])
len(data)
```

```python
#method for reshape all data into same shape
def vectorize_data(data,labels):
    all_data = np.zeros((len(data),data[0].shape[0],data[0].shape[1],data[0].shape[2]))
    all_labels = np.zeros((len(labels),data[0].shape[0]))
    for i in range(all_data.shape[0]):
        if data[i].shape[0] > 15 or len(labels[i]) > 15:
            all_data[i][:,:,:data[i].shape[2]] = data[i][:15,:,:]
            all_labels[i][:15] = labels[i][:15]
        elif data[i].shape[0] < 15 or len(labels[i]) < 15:
            all_data[i][:data[i].shape[0],:,:data[i].shape[2]] = data[i][:,:,:]
            all_labels[i,:len(labels[i])] = labels[i]
        else:
            all_data[i][:,:,:data[i].shape[2]] = data[i]
            all_labels[i] = labels[i][:15]
    return all_data,all_labels
```

```python
#write dataset and subjects of npz file for each subjects
def write_npz_file(x,y,folder,subject):
    if not os.path.exists(folder):
        os.makedirs(folder)
    filename = str(folder)+"/S"+str(subject).zfill(3)+".npz"
    np.savez(filename,x=x,y=y)
```

```python
x, y = vectorize_data(data,labels)
print(x.shape)
print(y.shape)
```

```python
#each subject have 3 runs, so reshape both dataset and label to have shape[1] = 3.
re_x = x.reshape(-1,3,15,64,481)
re_x = np.swapaxes(re_x,3,4)
print(re_x.shape)
re_y = y.reshape(-1,3,15)
print(re_y.shape)
```

```python
#write file
subject = 1
for i in range(re_x.shape[0]):
    if subject == 88 or subject == 92 or subject == 99:
        subject += 1
    write_npz_file(re_x[i],re_y[i],"lr_imagine",subject)
    subject += 1
```
