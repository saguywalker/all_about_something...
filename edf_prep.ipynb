```python
import numpy as np
import pyedflib
import pandas as pd
import matplotlib as plt
import scipy
import sklearn
```

```python
path_root = "subjects/"
def fetch_data(file_name):
    f = pyedflib.EdfReader(file_name)
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()
    sigbufs = np.zeros((n,f.getNSamples()[0]))
    for i in np.arange(n):
        sigbufs[i,:] = f.readSignal(i)
    return sigbufs
```

```python
#try the new ways
def fetch_data_anno(file_name):
    f = pyedflib.EdfReader(file_name)
    anno = f.readAnnotations()
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()
    sigbufs = np.zeros((n,f.getNSamples()[0]))
    for i in np.arange(n):
        sigbufs[i,:] = f.readSignal(i)
    return [sigbufs,anno]
```

```python
new_data = []
for i in range(1,110):
    folder_name = "subjects/S"+str(i).zfill(3)+"/S"+str(i).zfill(3)+"R"
    for j in range(1,15):
        full_path = folder_name+str(j).zfill(2)+".edf"
        new_data.append(fetch_data(full_path))
```
