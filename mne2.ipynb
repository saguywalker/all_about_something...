```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import ShuffleSplit, cross_val_score
from mne import Epochs, pick_types, find_events
from mne.channels import read_layout
from mne.io import concatenate_raws, read_raw_edf
from mne.datasets import eegbci
from mne.decoding import CSP
```

```python
tmin, tmax = -1.,1.
path_root = "subjects/"
```

```python
raw_files = []
for i in range(1,110):
    sub_folder = path_root+"S"+str(i).zfill(3)+"/S"+str(i).zfill(3)+"R"
    for j in range(1,15):
        full_path = sub_folder+str(j).zfill(2)+".edf"
        raw_files.append(read_raw_edf(full_path,preload=True,stim_channel="auto"))
```

```python
raw_events = []
for r in raw_files:
    raw_events.append(r.find_edf_events())
```

```python
events = []
for r in raw_files:
    events.append(find_events(r,shortest_event=0,stim_channel='STI 014'))
```

```python
events_without_baseline = []
baseline = []
for i in range(len(events)):
    if i % 14 == 0 or i % 14 - 1 == 0:
        baseline.append(events[i])
    else:
        events_without_baseline.append(events[i])
print(len(events_without_baseline))
print(len(baseline))
```
